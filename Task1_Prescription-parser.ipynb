### Creating a Tuples Maker method
Create the tuples as given below by writing a function **tuples_maker(input_sigs, output_labels)** and returns **output** as given below

Input(s): 
- input_sigs
- output_lables

Output:

[[('for', 'FOR'),
  ('5', 'Duration'),
  ('to', 'TO'),
  ('6', 'DurationMax'),
  ('days', 'DurationUnit')], [second sentence], ...]

```python
def tuples_maker(inp, out):
    sample_data = []
    for words, tags in zip(inp, out):
        sentence_tuple = []
        for word, tag in zip(words, tags):
            # Check if the word is in all capitals, convert to lowercase if true
            if word.isupper():
                word = word.lower()
            sentence_tuple.append((word, tag))

        step_size = 5
        for i in range(0, len(words), step_size):
            start_index = len(tags) + i
            sentence_tuple += [(word, tag) for word, tag in zip(words[start_index:], tags)]

        sample_data.append(sentence_tuple)
    return sample_data
```

### Creating the triples_maker( ) for feature extraction
- input: tuples_maker_output

[[('for', 'IN', 'FOR'),
  ('5', 'CD', 'Duration'),
  ('to', 'TO', 'TO'),
  ('6', 'CD', 'DurationMax'),
  ('days', 'NNS', 'DurationUnit')], [second sentence], ... ]

```python
def triples_maker(whole_data):
    sample_data = []
    for sentence in whole_data:
        for word, tag in sentence:
            (word, pos_tag([word])[0][1], tag)

    sample_data = [[(word, pos_tag([word])[0][1], tag) for word, tag in sentence] for sentence in whole_data]
    return sample_data
```

### Creating the features extractor method (GIVEN as a BASELINE)
#### The features used are:
- SOS, EOS, lowercase, uppercase, title, digit, postag, previous_tag, next_tag
#### Feel free to include more features

```python
def token_to_features(doc, i):
    word = doc[i][0]
    postag = doc[i][1]

    # Common features for all words
    features = [
        'bias',
        'word.lower=' + word.lower(),
        'word[-3:]=' + word[-3:],
        'word[-2:]=' + word[-2:],
        'word.isupper=%s' % word.isupper(),
        'word.istitle=%s' % word.istitle(),
        'word.isdigit=%s' % word.isdigit(),
        'postag=' + postag
    ]

    # Features for words that are not
    # at the beginning of a document
    if i > 0:
        word1 = doc[i-1][0]
        postag1 = doc[i-1][1]
        features.extend([
            '-1:word.lower=' + word1.lower(),
            '-1:word.istitle=%s' % word1.istitle(),
            '-1:word.isupper=%s' % word1.isupper(),
            '-1:word.isdigit=%s' % word1.isdigit(),
            '-1:postag=' + postag1
        ])
    else:
        # Indicate that it is the 'beginning of a document'
        features.append('BOS') #Begin of sentence

    # Features for words that are not
    # at the end of a document
    if i < len(doc)-1:
        word1 = doc[i+1][0]
        postag1 = doc[i+1][1]
        features.extend([
            '+1:word.lower=' + word1.lower(),
            '+1:word.istitle=%s' % word1.istitle(),
            '+1:word.isupper=%s' % word1.isupper(),
            '+1:word.isdigit=%s' % word1.isdigit(),
            '+1:postag=' + postag1
        ])
    else:
        # Indicate that it is the 'end of a document'
        features.append('EOS')

    return features
```

### Running the feature extractor on the training data 
- Feature extraction
- Train-test-split

```python
# A function for extracting features in documents
def get_features(doc):
    return [token_to_features(doc, i) for i in range (len(doc))]

# A function fo generating the list of labels for each document
def get_labels(doc):
    return [label for (token, postag, label) in doc]

#[print (doc) for doc in sample data]
X = [get_features(doc) for doc in sample_data]
y = [get_labels(doc) for doc in sample_data]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

### Training the CRF model with the features extracted using the feature extractor method

```python
trainer = pycrfsuite.Trainer()

# Submit training data to the trainer
for feature, label in zip(X,y):
    trainer.append(feature, label)

# Set the parameters of the model

trainer.set_params({
    'max_iterations': 1000,
    'c1': 0.1,
    'c2': 0.01
    })

# Providing a file name as a parameter to the train function, such that
# the model will be saved to the file when training is finished
trainer.train('prescription_parser.model') 
```
